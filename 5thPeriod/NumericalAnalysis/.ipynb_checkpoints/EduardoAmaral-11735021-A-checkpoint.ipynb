## Funções Auxiliares

### Verificação de matriz SPD

```python
def ehSPD(A):
  A = np.array(A)
  return np.all(A == A.T) and np.all(np.linalg.eigvals(A) > 0)
```

#### a) Escreva um código que monte a matriz A para $ n = 1000 $

```python
def gerMatrizAb(n):
  A = np.zeros((n,n), dtype=np.float64)
  b = np.full((n), np.power(n, -4.0))
  
  A[0][0] = A[1][1] = 100
  A[0][1] = A[1][0] = A[1][2] = -4
  A[0][2] = A[1][3] = 1
  A[n-2][n-2] = A[n-1][n-1] = 100
  A[n-2][n-1] = A[n-2][n-3] = A[n-1][n-2] = -4
  A[n-1][n-3] = A[n-2][n-4] = 1

  for i in range(2,n-2):
    A[i][i] = 100
    A[i][i+1] = A[i][i-1] = -4
    A[i][i+2] = A[i][i-2] = 1

  return A,b

A,b = gerMatrizAb(1000)

print(f'Matriz A = {A}')
print(f'Vetor b = {b}')
```

####b) Escreva um código que faça a decomposição de Cholesky de uma matriz simétrica definida positiva qualquer.

```python
def decomposicaoCholesky(A):
    if not ehSPD(A):
      raise ValueError('Matriz A não é simétrica definida positiva (SPD).')

    A = np.array(A, dtype=np.float64)
    n = len(A)
    H = np.tril(A)

    for i in range(n-1):
      H[i][i] = np.sqrt(H[i][i])
      H[i+1:n, i] = H[i+1:n, i] / H[i][i]
      for j in range(i+1,n):
        H[j:n, j] = H[j:n, j] - H[j:n, i] * H[j][i]

    H[n-1][n-1] = np.sqrt(H[n-1][n-1])

    return H

A = [[3,2,0], [2,5,1], [0,1,3]]
H = decomposicaoCholesky(A)

print(f'H = {H}')

print(f'A = H * H\'? {np.allclose(A, ( H * np.matrix(H).H), atol=1e-23)}')
```

#### c) Usando as rotinas implementadas anteriormente, escreva um código para resolver um sistema $Ax = b$.

```python
def substituicaoRegressiva(U, y):
  U = np.array(U, dtype=np.float64)
  y = np.array(y, dtype=np.float64)

  n = len(y)
  x = np.zeros(n)

  for i in reversed(range(n)):
    x[i] = (y[i] - U[i][i:n] @ x[i:n]) / U[i][i]

  return x

def substituicaoProgressiva(L, b):
  L = np.array(L, dtype=np.float64)
  b = np.array(b, dtype=np.float64)
  n = len(b)
  x = np.zeros(n)

  for i in range(n):
    x[i] = (b[i] - L[i][:i] @ x[:i]) / L[i][i]
  
  return x

def resolverUsandoCholesky(A,b):
  A = np.array(A, dtype=np.float64)
  b = np.array(b, dtype=np.float64)

  L = decomposicaoCholesky(A)

  y = substituicaoProgressiva(L, b)

  x = substituicaoRegressiva(L.T, y)

  return x

A,b = gerMatrizAb(1000)

x = resolverUsandoCholesky(A,b)

print(f'A solução do sistema por meio da decomposição de Cholesky está correta? {np.isclose(np.linalg.norm(A@x - b),0, atol= 1e-23)}')
```

#### d) Escreva um código implementando o método de Jacobi para resolver um sistema $Ax = b$. Utilize o erro absoluto como critério de parada.

```python
def jacobiConverge(A):
    A = np.array(A, dtype=np.float64)
    n = len(A)
    for i in range(n):
        somaAbs = sum(abs(A[i][j]) for j in range(n) if j != i)
        if abs(A[i][i]) <= somaAbs:
            return False
    return True

def resolverUsandoJacobi(A,b, x0=None, tol=1e-10, maxK=None):

  if not jacobiConverge(A):
      raise ValueError('Matriz A não atende os critérios de convergência do método.')

  A = np.array(A, dtype=np.float64)
  b = np.array(b, dtype=np.float64)  

  n = len(A);

  if x0 is None:
    x0 = np.zeros(n)

  dInv = np.zeros((n,n), dtype=np.float64);

  np.fill_diagonal(dInv, 1 / np.diag(A));

  I = np.eye(n);

  C = I - np.dot(dInv, A);

  g = np.dot(dInv, b);

  k = 0

  while np.linalg.norm(b-np.dot(A,x0)) > tol and (maxK is None or k < maxK):
    x0 = np.dot(C, x0) + g;
    k+=1
  
  return x0, k

A,b = gerMatrizAb(1000)

x,k = resolverUsandoJacobi(A,b, tol=1e-23)

print(f'A solução do sistema por meio do método Jacobi está correta? {np.isclose(np.linalg.norm(A@x - b),0, atol= 1e-23)}')
print(f'A solução demorou {k} iterações')
```

#### e) Vamos comparar o método direto de Cholesky com o iterativo de Jacobi neste exemplo. Observe quanto tempo leva para resolver o sistema usando Cholesky. Quantas iterações foram necessárias no método de Jacobi para obtermos a mesma precisão da solução dada pelo método de Cholesky ?

```python
A,b = gerMatrizAb(1000)

inicioCholesky = time.time()
xCholesky = resolverUsandoCholesky(A,b)
fimCholesky = time.time()
tempoCholesky = fimCholesky - inicioCholesky

inicioJacobi = time.time()
xJacobi, kJacobi = resolverUsandoJacobi(A,b, tol=1e-23)
fimJacobi = time.time()
tempoJacobi = fimJacobi - inicioJacobi

print(f'Ambos os métodos chegaram no mesmo resultado? {np.allclose(xCholesky, xJacobi, atol=1e-23)}')

print(f'Tempo gasto pelo método Cholesky: {tempoCholesky} s')

print(f'Tempo gasto pelo método Jacobi: {tempoJacobi} s')

print(f'Número de iterações gastas pelo método Jacobi para a mesma precisão do Cholesky: {kJacobi}')
```

### 1) Observando a equação (4), vê-se a necessidade de calcular a matriz inversa da Jacobiana. É possı́vel calcular matrizes inversas usando a decomposição LU, sendo assim, implemente um código que:

#### a) Calcule a fatoração LU de uma matriz qualquer

```python
def fatoracaoLU(A):
  A = np.array(A, dtype=np.float32)
  n = len(A)
  L = np.eye(n)
  U = np.zeros((n,n))

  for k in range(0,n):
    for j in range(k,n):
      U[k][j] = A[k][j] - (np.dot(L[k][0:k+1],U[0:k+1,j]))
    for i in range(k+1,n):
      L[i][k] = (A[i][k] - np.dot(L[i][0:k],U[0:k,k]))/U[k][k]
  
  return (L,U)
      


A = np.array([[2,1,2],[4,3,3],[6,5,-1]])
L,U = fatoracaoLU(A)

print(L)
print(U)

print(f'A = L x U? {np.allclose(A, L @ U, atol=1e-10)}')
```

#### b) Resolva um sistema linear cujo lado direito é uma matriz



```python
def resolverUsandoLU(A,b):
  A = np.array(A, dtype=np.float64)
  b = np.array(b, dtype=np.float64)

  L,U = fatoracaoLU(A)

  n = len(A)

  if len(b.shape) == 1:
    y = substituicaoProgressiva(L, b)

    x = substituicaoRegressiva(U, y)

    return x

  m = b.shape[1]

  X = np.zeros((n,m), dtype=np.float64)

  for k in range(m):
    y = substituicaoProgressiva(L, b[:,k])

    x = substituicaoRegressiva(U, y)

    X[:,k] = x

  return X

np.random.seed(42)

A = np.random.randint(1,1000, (10,10))
b = np.random.randint(1,1000, (10,10))

x = resolverUsandoLU(A,b)

print(f'A solução do sistema por meio da fatoração LU está correta? {np.isclose(np.linalg.norm(A @ x - b),0, atol=1e-10)}')
```

#### c) Finalmente, calcule a inversa de uma matriz qualquer resolvendo vários sistemas lineares

```python
def inversaPorLU(A):
  n = len(A)
  I = np.eye(n)
  
  inversa = resolverUsandoLU(A, I)
  return inversa

np.random.seed(42)

A = np.random.randint(1,1000, (10,10))

invA = inversaPorLU(A)

print(f'O calculo da inversa por meio da resolução de sistemas funcionou? {np.allclose(A @ invA, np.eye(len(A)), atol=1e-10)}')
```

### 2) Implemente o método de Newton para sistemas usando a rotina implementada no item anterior. Lembre-se de especificar o critério de parada utilizado.


```python
def resolverUsandoNewton(F, J, nEquations, x0=None, maxK=None, tol=1e-10):
    if x0 is None:
      x0 = np.ones(nEquations)

    k = 0

    while maxK is None or k < maxK:
        valF = F(x0)
        invJ = inversaPorLU(J(x0))
        dX = invJ @ valF
        x0 -= dX
        if np.linalg.norm(dX) < tol:
            return x0,k
        k+=1
    return None

def F(x):
    f1 = 3 * x[0] - x[1] - 89
    f2 = -x[0] + 4 * x[1] + 32
    return np.array([f1, f2])

def J(x):
    j11 = 3
    j12 = -1
    j21 = -1
    j22 = 4
    return np.array([[j11, j12], [j21, j22]])

x, k = resolverUsandoNewton(F, J, 2)

print(f'A solução encontrada utilizando o método de Newton está correta? {np.allclose(F(x), 0, atol=1e-10)}')
print(f'O método convergeu após {k} iterações')
```

Calcular inversas de matrizes é um processo que demanda muito custo computacional (como pode ser observado nos itens anteriores, precisa-se de uma fatoração LU e da resolução de vários sistemas lineares). Por isso, ao usar o método de Newton para sistemas, faz-se algumas manipulações na equação (4) evitando o cálculo da inversa da matriz Jacobiana. Define-se um vetor $z = x^{k+1} −x^k$ e resolve-se o sistema linear $J_F(x^k )z = F(x^k)$. Deste modo, o algoritmo para resolver um sistema de equações não-lineares usando o método de Newton é :

1. Numa dada iteração $k$, resolve-se o sistema $J_F(x^k)z = F(x^k)$,
2. Atualiza-se $x^{k+1} = x^k − z$.

### 3) Implemente novamente o método de Newton para sistemas usando o algoritmo anterior. Utilizeas rotinas já implementadas neste trabalho para resolução do sistema linear, indicando sempre qual está usando.

```python
def resolverUsandoNewtonComZ(F, J, nEquations, x0=None, maxK=None, tol=1e-10, systemSolver = np.linalg.solve):
    if x0 is None:
      x0 = np.ones(nEquations)

    k = 0

    while maxK is None or k < maxK:
        valF = F(x0)
        valJ = J(x0)
        systemRes = systemSolver(valJ, -valF)
        z = systemRes if type(systemRes) is not tuple else systemRes[0]
        x0 += z
        if np.linalg.norm(z) < tol:
            return x0,k
        k+=1
    return None

def F(x):
    f1 = 3 * x[0] - x[1] - 89
    f2 = -x[0] + 4 * x[1] + 32
    return np.array([f1, f2])

def J(x):
    j11 = 3
    j12 = -1
    j21 = -1
    j22 = 4
    return np.array([[j11, j12], [j21, j22]])

x, k = resolverUsandoNewtonComZ(F, J, 2, systemSolver=resolverUsandoLU)

print(f'A solução encontrada utilizando o método de Newton com Z e solução de sistemas por LU está correta? {np.allclose(F(x), 0, atol=1e-10)}')
print(f'O método convergeu após {k} iterações')

x, k = resolverUsandoNewtonComZ(F, J, 2, systemSolver=resolverUsandoJacobi)

print(f'A solução encontrada utilizando o método de Newton com Z e solução de sistemas por Jacobi está correta? {np.allclose(F(x), 0, atol=1e-10)}')
print(f'O método convergeu após {k} iterações')

# Não se testa com o método de Cholesky pois a matriz gerada não é SPD
```

### 4) Dada a equação de um cı́rculo $(x − a)^2 + (y − b)^2 = R^2$ e três pontos que passam por esse cı́rculo

image.png

#### a) Monte um sistema não-linear para determinar $a$, $b$ e $R$.

**Equação 1:** $(8.21 - a)^2 + (-b)^2 = R^2 $

**Equação 2:** $(0.34 - a)^2 + (6.62 - b)^2 = R^2 $

**Equação 3:** $(5.96 - a)^2 + (-1.12 - b)^2 = R^2 $

#### b) Resolva o sistema não-linear utilizando os códigos feitos nos itens **2** e **3**.

```python
def FCirc(ps, x, y):
    a, b, R = ps
    f1 = (x[0] - a)**2 + (y[0] - b)**2 - R**2
    f2 = (x[1] - a)**2 + (y[1] - b)**2 - R**2
    f3 = (x[2] - a)**2 + (y[2] - b)**2 - R**2
    return np.array([f1, f2, f3])

def JCirc(ps, x, y):
    a, b, R = ps
    jacobiano = np.zeros((3, 3))
    jacobiano[0, 0] = -2 * (x[0] - a)
    jacobiano[0, 1] = -2 * (y[0] - b)
    jacobiano[0, 2] = -2 * R
    jacobiano[1, 0] = -2 * (x[1] - a)
    jacobiano[1, 1] = -2 * (y[1] - b)
    jacobiano[1, 2] = -2 * R
    jacobiano[2, 0] = -2 * (x[2] - a)
    jacobiano[2, 1] = -2 * (y[2] - b)
    jacobiano[2, 2] = -2 * R
    return jacobiano

def determinarParametrosCirculo(xs, ys, nonLinearSystemSolver=resolverUsandoNewton, systemSolver=None):
    if systemSolver is None:
      (a,b,R), k = nonLinearSystemSolver(lambda ps: FCirc(ps, xs, ys), lambda ps: JCirc(ps, xs, ys), 3)
    else:
      (a,b,R), k = nonLinearSystemSolver(lambda ps: FCirc(ps, xs, ys), lambda ps: JCirc(ps, xs, ys), 3, systemSolver=systemSolver)
    return (a,b,R), k

xs = (8.21, 0.34, 5.96)
ys = (0, 6.62, -1.12)

(a,b,R), k = determinarParametrosCirculo(xs,ys, nonLinearSystemSolver=resolverUsandoNewton)

print(f'A solução encontrada utilizando o método de Newton está correta? {np.allclose(FCirc((a,b,R),xs,ys), 0, atol=1e-10)}')
print(f'Solucão a = {a}, b = {b}, R = {R}')
print(f'O método convergeu após {k} iterações')

(a,b,R), k = determinarParametrosCirculo(xs,ys, nonLinearSystemSolver=resolverUsandoNewtonComZ, systemSolver=resolverUsandoLU)

print(f'A solução encontrada utilizando o método de Newton e solução de sistemas por LU está correta? {np.allclose(FCirc((a,b,R),xs,ys), 0, atol=1e-10)}')
print(f'Solucão a = {a}, b = {b}, R = {R}')
print(f'O método convergeu após {k} iterações')

# Não se testa com o método de Cholesky pois a matriz gerada não é SPD
# Não se testa com o método de Jacobi pois a matriz gerada não atende os critérios de convergência
```

#### c) Houve melhora no tempo de execução do código implementado no item **3**?

```python
tsNewton = []

for i in range(1000):
  inicioNewton = time.time()
  resNewton, _ = determinarParametrosCirculo(xs,ys, nonLinearSystemSolver=resolverUsandoNewton)
  fimNewton = time.time()
  tempoNewton = fimNewton - inicioNewton
  tsNewton.append(tempoNewton)

tsNewtonComZLU = []

for i in range(1000):
  inicioNewtonComZLU = time.time()
  resNewtonComZLU, _ = determinarParametrosCirculo(xs,ys, nonLinearSystemSolver=resolverUsandoNewtonComZ, systemSolver=resolverUsandoLU)
  fimNewtonComZLU = time.time()
  tempoNewtonComZLU = fimNewtonComZLU - inicioNewtonComZLU
  tsNewtonComZLU.append(tempoNewtonComZLU)

print(f'Ambos os métodos chegaram no mesmo resultado? {np.allclose(resNewton, resNewtonComZLU, atol=1e-23)}')

print(f'Tempo médio gasto pelo método de Newton com a inversa do Jacobiano em 1000 execuções: {np.mean(tsNewton)} s')

print(f'Tempo médio gasto pelo método de Newton com Z usando o método LU para resolver sistemas em 1000 execuções: {np.mean(tsNewtonComZLU)} s')


```

É possível perceber uma pequena melhora ao utilizar o método de Newton com uso do vetor $Z$, porém a diferenca é bem pequena. Talvez ao utilizar outros métodos para resolução de sistemas a melhora seja ainda maior, entretanto, no exercício em questão a matriz gerada não é SPD, o que impede o uso do método de Cholesky e não atende os critérios de convergência de Jacobi, o que impede o uso do método.

```python
xs = (8.21, 0.34, 5.96)
ys = (0, 6.62, -1.12)

(a,b,R), _= determinarParametrosCirculo(xs,ys, nonLinearSystemSolver=resolverUsandoNewtonComZ, systemSolver=resolverUsandoLU)

x = np.linspace(-15, 15, 1000)
y = np.linspace(-15, 15, 1000)

X, Y = np.meshgrid(x,y)

f = (X-a)**2 + (Y-b)**2 - R**2

plt.title(f'( x - {np.round(a,2)} )² + ( y - {np.round(b,2)} )² - {np.round(R,2)}² = 0')

plt.contour(X,Y,f,[0])

for x,y in zip(xs,ys):
  plt.plot(x, y, "or")
  plt.annotate(f'({x},{y})', (x,y))

plt.show()
```
