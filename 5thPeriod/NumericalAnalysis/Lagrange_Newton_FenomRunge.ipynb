```python
import numpy as np
from scipy import linalg as lin
import scipy as sci
import math as m
import matplotlib.pyplot as plt
from sympy import *
import time


"""# Interpolao

## Interpolao de Lagrange
"""

def lagrange_interp(xi,yi,x):
    n = np.size(xi); # maior dimenso de um vetor unidimensional
    m = np.size(x); # maior dimenso de um vetor unidimensional
    
    L = np.ones((n,m));

    for i in np.arange(n):
        for j in np.arange(n):
            if(i != j):
                L[i,:] = (L[i,:]*(x-xi[j]))/(xi[i]-xi[j]);

    y = yi.dot(L); # --> yi  vetor linha 1xn L  matriz nxm --> 1xm
    return y;

# Exemplo slide 12 interpolation

xi = np.array([-2,0,3,5], dtype='double');
print(xi);
yi = np.array([3,-2,4,2], dtype='double');
print(yi);
x = np.linspace(-2, 5, num=41, endpoint=True)

y = lagrange_interp(xi,yi,x)

plt.figure(figsize=(10,6),facecolor='white')
plt.plot(x,y,label = 'Interpolao de Lagrange',linewidth = 2)
plt.scatter(xi,yi,label = 'Pontos',linewidth = 2) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('') 
plt.legend() 
plt.show()

# Interpolando a funo cos(x) com um polinmio de grau 9 (temos que usar 10 pontos) no intervalo (0,1)

xi = np.linspace(0, 1, num=10, endpoint=True)
print(xi);
yi = np.cos(xi);
print(yi)
x = np.linspace(0, 1, num=101, endpoint=True)
print(np.size(x))
y = lagrange_interp(xi,yi,x);

plt.figure(figsize=(10,6),facecolor='white')
plt.plot(x,y,label = 'Interpolao de Lagrange',linewidth = 2,zorder=1)
plt.plot(x,np.cos(x),label = 'Cosseno',linewidth = 2,linestyle='--',zorder=2)
plt.scatter(xi,yi,label = 'Pontos',linewidth = 3,facecolor='red',zorder=3) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('') 
plt.legend() 
plt.show()

"""## Forma de Newton"""

def newton_interp(xi,yi,x):
    n = np.size(xi); ni = np.size(x); N = np.ones((n,ni));
    D = np.zeros((n,n)); D[:,0] = yi;
    
    for j in np.arange(n-1): # matriz de diferenas divididas
        for i in np.arange(n-j-1):
            D[i,j+1] = (D[i+1,j]-D[i,j])/(xi[i+j+1]-xi[i]);

    for i in np.arange(1,n): # loop do produtrio da forma de Newton
        N[i,:] = N[i-1,:]*(x-xi[i-1]);
    
    y = D[0,:].dot(N)
    
    return y

# Exemplo slide 12 interpolation

xi = np.array([-2,0,3,5], dtype='double');
print(xi);
yi = np.array([3,-2,4,2], dtype='double');
print(yi);
x = np.linspace(-2, 5, num=41, endpoint=True)

y = newton_interp(xi,yi,x)

plt.figure(figsize=(10,6),facecolor='white')
plt.plot(x,y,label = 'Forma de Newton',linewidth = 2)
plt.scatter(xi,yi,label = 'Pontos',linewidth = 2) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('') 
plt.legend() 
plt.show()

# Interpolando a funo cos(x) com um polinmio de grau 9 (temos que usar 10 pontos) no intervalo (0,1)

xi = np.linspace(0, 1, num=10, endpoint=True)
print(xi);
yi = np.cos(xi);
print(yi)
x = np.linspace(0, 1, num=101, endpoint=True)
print(np.size(x))
y = newton_interp(xi,yi,x);

plt.figure(figsize=(10,6),facecolor='white')
plt.plot(x,y,label = 'Forma de Newton',linewidth = 2,zorder=1)
plt.plot(x,np.cos(x),label = 'Cosseno',linewidth = 2,linestyle='--',zorder=2)
plt.scatter(xi,yi,label = 'Pontos',linewidth = 3,facecolor='red',zorder=3) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('') 
plt.legend() 
plt.show()

"""## Comparando as interpolaes

Vamos pegar os exemplos anteriores e fazer um grfico com os resultados de Lagrange e Newton.
"""

# Exemplo slide 12 interpolation

xi = np.array([-2,0,3,5], dtype='double');
print(xi);
yi = np.array([3,-2,4,2], dtype='double');
print(yi);
x = np.linspace(-2, 5, num=41, endpoint=True)

y_lagrange = lagrange_interp(xi,yi,x)
y_newton = newton_interp(xi,yi,x)

plt.figure(figsize=(10,6),facecolor='white')
plt.plot(x,y_lagrange,label = 'Interpolao de Lagrange',linewidth = 2,zorder=2)
plt.plot(x,y_newton,label = 'Forma de Newton',linewidth = 2,marker='>',zorder=1)
plt.scatter(xi,yi,label = 'Pontos',linewidth = 2,facecolor='red',zorder=3) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('') 
plt.legend() 
plt.show()

# Interpolando a funo cos(x) com um polinmio de grau 9 (temos que usar 10 pontos) no intervalo (0,1)

xi = np.linspace(0,1,num=10, endpoint=True)
print(xi);
yi = np.cos(xi);
print(yi)
x = np.linspace(0,1,num=101, endpoint=True)
print(np.size(x))

y_lagrange = lagrange_interp(xi,yi,x)
y_newton = newton_interp(xi,yi,x);

plt.figure(figsize=(10,6),facecolor='white')
plt.plot(x,y_lagrange,label = 'Interpolao de Lagrange',linewidth = 2,marker='o',zorder=2)
plt.plot(x,y_newton,label = 'Forma de Newton',linewidth = 2,marker='>',zorder=1)
plt.plot(x,np.cos(x),label = 'Cosseno',linewidth = 2,linestyle='--',zorder=3)
plt.scatter(xi,yi,label = 'Pontos',linewidth = 2,facecolor='red',zorder=4) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('') 
plt.legend() 
plt.show()

"""Agora vamos observar o fenmeno de Runge com a funo
\begin{equation}
    f(x) = \dfrac{1}{1+25x^2}, \; x \in [-1,1]
\end{equation}
"""

# 5 ns de interpolao
xi = np.linspace(-1,1,num=5, endpoint=True)
# print(xi)
yi = 1/(1+25*xi**2);
# print(yi)
x = np.linspace(-1,1,num=60, endpoint=True)
y_exat = 1/(1+25*x**2);

y_lagrange5 = lagrange_interp(xi,yi,x);
y_newton5 = newton_interp(xi,yi,x)


# 7 ns de interpolao
xi = np.linspace(-1,1,num=7, endpoint=True)
# print(xi)
yi = 1/(1+25*xi**2);
# print(yi)
x = np.linspace(-1,1,num=60, endpoint=True)
y_exat = 1/(1+25*x**2);

y_lagrange7 = lagrange_interp(xi,yi,x);
y_newton7 = newton_interp(xi,yi,x)


# 16 ns de interpolao
xi = np.linspace(-1,1,num=16, endpoint=True)
# print(xi)
yi = 1/(1+25*xi**2);
# print(yi)
x = np.linspace(-1,1,num=60, endpoint=True)
y_exat = 1/(1+25*x**2);

y_lagrange16 = lagrange_interp(xi,yi,x);
y_newton16 = newton_interp(xi,yi,x)

f = plt.figure(figsize=(20,6),facecolor='white')
f.add_subplot(1,3,1);
plt.plot(x,y_lagrange5,label = 'Interpolao de Lagrange',linewidth = 2,linestyle='--',zorder=2)
plt.plot(x,y_newton5,label = 'Forma de Newton',linewidth = 2,marker='>',zorder=1)
plt.plot(x,y_exat,label = 'Funo Original',linewidth = 2,zorder=3)
plt.legend()
plt.title('5 pontos') 
f.add_subplot(1,3,2);
plt.plot(x,y_lagrange7,label = 'Interpolao de Lagrange',linewidth = 2,linestyle='--',zorder=2)
plt.plot(x,y_newton7,label = 'Forma de Newton',linewidth = 2,marker='>',zorder=1)
plt.plot(x,y_exat,label = 'Funo Original',linewidth = 2,zorder=3)
plt.legend()
plt.title('7 pontos')
f.add_subplot(1,3,3);
plt.plot(x,y_lagrange16,label = 'Interpolao de Lagrange',linewidth = 2,linestyle='--',zorder=2)
plt.plot(x,y_newton16,label = 'Forma de Newton',linewidth = 2,marker='>',zorder=1)
plt.plot(x,y_exat,label = 'Funo Original',linewidth = 2,zorder=3)
plt.legend()
plt.title('16 pontos')
# plt.scatter(xi,yi,label = 'Pontos',linewidth = 2,facecolor='red',zorder=4) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large')  
plt.show()

"""## Ns de Chebyshev"""

a = -1; b = 1;
# 17 ns de interpolao
n = 16;
xi = np.zeros(n+1);
i = np.arange(n+1);
xi = (a+b)/2 + (b-a)*np.cos((2*i+1)*m.pi/(2*(n+1)))/2
# print(xi)

yi = 1/(1+25*xi**2);
# print(yi)
x = np.linspace(-1,1,num=60, endpoint=True)
y_exat = 1/(1+25*x**2);

y_lagrange5 = lagrange_interp(xi,yi,x);
y_newton5 = newton_interp(xi,yi,x)

plt.figure(figsize=(10,6),facecolor='white')
plt.plot(x,y_lagrange5,label = 'Interpolao de Lagrange',linewidth = 2,linestyle='--',zorder=2)
plt.plot(x,y_newton5,label = 'Forma de Newton',linewidth = 2,marker='>',zorder=1)
plt.plot(x,y_exat,label = 'Funo Original',linewidth = 2,zorder=3)
plt.scatter(xi,yi,label = 'Pontos',linewidth = 2,facecolor='red',zorder=4) 
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('16 pontos') 
plt.legend() 
plt.show()
```
