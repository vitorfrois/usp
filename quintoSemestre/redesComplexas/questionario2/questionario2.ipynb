```python
import networkx as nx
import statistics as stat
import numpy as np
```

```python
G = nx.erdos_renyi_graph(100, 0.2)
nx.draw(G)
```

```python
def compute_spl_list(G):
  spl_list = []
  for source_node in nx.nodes(G):
    for target_node in nx.nodes(G):
      distance = nx.shortest_path_length(G, source=source_node, target=target_node)
      spl_list.append(distance)
  spl_list = np.array(spl_list)
  return spl_list
```

```python
n = G.number_of_nodes()
den = n * (n - 1)
spl_list = compute_spl_list(G)
avg_spl = spl_list.sum()/den
var_spl = spl_list.var()
print(avg_spl)
print(var_spl)
```
